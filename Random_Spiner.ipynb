```python
import random
import matplotlib.pyplot as plt

def spinner(probability_distribution):
    return random.choices(range(len(probability_distribution)), weights=probability_distribution)[0]

def draw_spinner(probability_distribution, labels):
    fig, ax = plt.subplots()
    ax.pie(probability_distribution, labels=labels, autopct='%1.1f%%', startangle=90, colors=['blue', 'orange', 'yellow', 'green'])
    ax.axis('equal')  # AsigurÄƒ un aspect circular
    plt.show()

probabilities = [0.1286, 0.2429, 0.2, 0.4286] 
labels = ['45 degrees', '85 degrees', '70 degrees', '150 degrees']
selected_index = spinner(probabilities)
print(f"Indexul selectat: {selected_index}")

draw_spinner(probabilities, labels)
```
